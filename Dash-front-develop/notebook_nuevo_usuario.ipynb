```python
nuevo_usuario = ratings['userid'].max().split("_")
nuevo_usuario = ("user_" + str(1000000 + int(nuevo_usuario[1])+1)).replace("_1", "_")
nuevo_usuario
```

```python
np.random.seed(len(ratings.groupby('userid').size()))

top_100 = rep_artista.sort_values('rating_count', ascending=False).head(100)
top_100['aleatorio'] = np.random.randint(0,100, 100)
top_100 = top_100.sort_values('aleatorio')
top_100['aleatorio'] = list(range(1, 11))*10
top_100
```

```python
eleccion = []
grupo = 1

while(len(eleccion)<5):
    z = top_100[top_100['aleatorio'] == grupo].reset_index()
    z['ARTISTA_NUM'] = range(1,11)
    z = z[['ARTISTA_NUM', 'artid']]
    print(z)
    ingreso = input("Ingresa los artistas que te gustan (separados por comas). Si ninguno te gusta, escribe 0")
    if ingreso != "0":
        for i in ingreso.split(","):
            eleccion.append(z[z['ARTISTA_NUM']==int(i)])
```

```python
z = pd.DataFrame()

for i in range(len(eleccion)):
    x = pd.DataFrame(eleccion[i])
    z = z.append(x)
    z['userid'] = nuevo_usuario
```
